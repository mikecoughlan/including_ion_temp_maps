```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
desc_dict = {'grl_0':{}, 'grl_1':{}, 'grl_2':{}, 'grl_3':{}, 'svlb':{}}

desc_dict['grl_0']['train'] = pd.read_feather('outputs/GRL-0_swmag_v6-1_train_description.feather')
desc_dict['grl_0']['val'] = pd.read_feather('outputs/GRL-0_swmag_v6-1_val_description.feather')
desc_dict['grl_0']['test'] = pd.read_feather('outputs/GRL-0_swmag_v6-1_test_description.feather')

desc_dict['grl_1']['train'] = pd.read_feather('outputs/GRL-1_swmag_v6-1_train_description.feather')
desc_dict['grl_1']['val'] = pd.read_feather('outputs/GRL-1_swmag_v6-1_val_description.feather')
desc_dict['grl_1']['test'] = pd.read_feather('outputs/GRL-1_swmag_v6-1_test_description.feather')

desc_dict['grl_2']['train'] = pd.read_feather('outputs/GRL-2_swmag_v6-1_train_description.feather')
desc_dict['grl_2']['val'] = pd.read_feather('outputs/GRL-2_swmag_v6-1_val_description.feather')
desc_dict['grl_2']['test'] = pd.read_feather('outputs/GRL-2_swmag_v6-1_test_description.feather')

desc_dict['grl_3']['train'] = pd.read_feather('outputs/GRL-3_swmag_v6-1_train_description.feather')
desc_dict['grl_3']['val'] = pd.read_feather('outputs/GRL-3_swmag_v6-1_val_description.feather')
desc_dict['grl_3']['test'] = pd.read_feather('outputs/GRL-3_swmag_v6-1_test_description.feather')

desc_dict['svlb']['train'] = pd.read_feather('outputs/SVLB_swmag_v6-1_train_description.feather')
desc_dict['svlb']['val'] = pd.read_feather('outputs/SVLB_swmag_v6-1_val_description.feather')
desc_dict['svlb']['test'] = pd.read_feather('outputs/SVLB_swmag_v6-1_test_description.feather')
```

```python
placements = [1, 7, 13]
colors = ['r', 'g', 'b', 'y', 'c']
mag_params = ['dbht_median', 'MAGNITUDE_median', 'MAGNITUDE_std', 'sin_theta_std', 'cos_theta_std', 'cosMLT', 'sinMLT']

stat = 'mean'
# var = 'MAGNITUDE_std'

# plotting the stats of the dataset for each region next to each other for comparison
fig, axs = plt.subplots(4, 2, figsize=(15, 15))
for i, var in enumerate(mag_params):
	xPlot = i // 2
	yPlot = i % 2
	axs[xPlot,yPlot].set_title(f'{var}_{stat}')
	for k, grl in enumerate(desc_dict.keys()):
		for l, subset in enumerate(desc_dict[grl].keys()):
			bar_loc = placements[l]+k
			axs[xPlot,yPlot].bar(bar_loc, desc_dict[grl][subset].loc[stat, var], bottom=0, color=colors[k], label=f'{grl}_{subset}')
			axs[xPlot,yPlot].set_ylabel(stat)
			axs[xPlot,yPlot].set_xticks([3, 9, 15])
			axs[xPlot,yPlot].set_xticklabels(desc_dict['grl_0'].keys())
```
